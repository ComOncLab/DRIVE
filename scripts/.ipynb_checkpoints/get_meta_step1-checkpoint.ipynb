```python
import pysradb
from pysradb.sraweb import SRAweb
db = SRAweb()
```

```python
import pandas as pd
```

```python
all_srr = pd.read_csv("/home/wt/GEO_data/SRA/curr_files",header=None,names=["srr"])
```

```python
all_srr
```

```python
from tqdm import tqdm
import sys 
import time
```

```python
for i in tqdm(range(len(all_srr))):
    try:
        df = db.sra_metadata(all_srr.srr[i], detailed=True,output_read_lengths=True)
        df.to_csv("metadata/"+str(all_srr.srr[i])+".txt",sep="\t", index=False)
    except:
        time.sleep(20)
        try:
            df = db.sra_metadata(all_srr.srr[i], detailed=True,output_read_lengths=True)
            df.to_csv("metadata/"+str(all_srr.srr[i])+".txt",sep="\t", index=False)
        except:
            time.sleep(20)
```
